**Aim: Implement Naive Bayes classifier : Whether Example**

Step 1: Import necessary libraries.

```python
from sklearn import preprocessing
from sklearn.naive_bayes import GaussianNB, MultinomialNB
import pandas as pd
```

Step 2: Prepare dataset.

```python

weather = ['Sunny', 'Sunny', 'Overcast', 'Rainy', 'Rainy','Rainy', 'Overcast',
           'Sunny', 'Sunny', 'Rainy', 'Sunny', 'Overcast', 'Overcast', 'Rainy']

temp = ['Hot','Hot','Hot','Mild','Cool','Cool','Cool','Mild',
        'Cool','Mild','Mild','Mild','Hot','Mild']

play=['No','No','Yes','Yes','Yes','No','Yes','No','Yes',
      'Yes','Yes','Yes','Yes','No']
```

```python
df = pd.DataFrame(list(zip(weather,temp,play)), columns=['Weather', 'Temperature', 'Play'])
df
```

Step 3: Digitize the data set using encoding

```python
#creating labelEncoder
d_lebal = preprocessing.LabelEncoder() # apply different label

# Converting string labels into numbers.
weather_encoded=d_lebal.fit_transform(weather)
print("Weather:" ,weather_encoded)
```

```python

temp_encoded_detail=d_lebal.fit_transform(temp)
label=d_lebal.fit_transform(play)

print("Temp:",temp_encoded_detail)
print("Play:",label)
```

Step 4: Merge different features to prepare dataset

```python
#Combinig weather and temp into single listof tuples
features=tuple(zip(weather_encoded,temp_encoded_detail))
print("Features:",features)
```

Step 5: Train ’Naive Bayes Classifier’

```python
#Create a Classifier
model=MultinomialNB()
# Train the model using the training sets
model.fit(features,label)
```

Step 6: Predict Output for new data

```python
#Predict Output
predicted= model.predict([[0,2]]) # 0:Overcast, 2:Mild
print("Predicted Value:", predicted)
```
