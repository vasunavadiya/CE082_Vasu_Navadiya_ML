```python
import numpy as np
import pandas as pd

matrix_3x2 = np.array([[1, 1], [2, 2], [3, 3]])
matrix_2x3 = np.array([[1, 1, 1], [2, 2, 3]])

print('3 x 2 matrix : \n', matrix_3x2)
print('\n 2 x 3 matrix : \n', matrix_2x3)
```

```python
random_matrix = np.random.rand(3, 4)
print(random_matrix)
```

```python
# matrix multiplication
matrix_mul = np.dot(matrix_3x2, matrix_2x3)
print(matrix_mul)
```

```python
# elemet wise matrix multiplication
ans= [[0 for x in range(len(matrix_3x2))] for y in range(len(matrix_2x3[0]))]

for i in range(len(matrix_3x2)): 
  for j in range(len(matrix_2x3[0])): 
    for k in range(len(matrix_2x3)): 
      ans[i][j] += matrix_3x2[i][k] * matrix_2x3[k][j]
 
print('Element wise multiplication : \n', ans)
```

```python
mean_1 = np.mean(matrix_3x2)
mean_2 = np.mean(matrix_2x3)
print(mean_1)
print(mean_2)
```
